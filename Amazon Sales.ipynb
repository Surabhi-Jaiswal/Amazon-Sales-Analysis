```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='rating', y='rating_count', data=df)
plt.title('Ratings vs Number of Reviews')
plt.show()
```

```python
top_reviewed_products = df[['product_name', 'rating_count']].sort_values(by='rating_count', ascending=False).head(10)
top_reviewed_products
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='category', data=df, order=df['category'].value_counts().index)
plt.title('Distribution of Products by Category')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['rating'], bins=5, kde=True)
plt.title('Distribution of Ratings')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['discount_percentage'], bins=10, kde=True)
plt.title('Distribution of Discount Percentages')
plt.show()
```
